```python
#DESCRIPTIONS FINAL WORKING CODE
liste2 = []
flat_list2 = []
descriptions = []
for i in biglist:
    liste2.append(i.select('div > .game_text'))
for i in liste2:
    if not i:
        i.append("null")
for sublist in liste2:
    for item in sublist:
        flat_list2.append(item)
for j in flat_list2:
    if j == 'null':
        descriptions.append("null")
    else:
        descriptions.append(j.get_text())
descriptions
```

```python
#SALE TAG FINAL WORKING CODE
liste3 = []
flat_list3 = []
saletag = []
for i in biglist:
    liste3.append(i.select('.sale_tag'))
for i in liste3:
    if not i:
        i.append("null")
for sublist in liste3:
    for item in sublist:
        flat_list3.append(item)
for j in flat_list3:
    if j == 'null':
        saletag.append("null")
    else:
        saletag.append(j.get_text())
saletag
```

```python
links = []
for div in soup.select('div > .thumb_link'):
    if div.has_attr('href'):
        links.append(div['href'])
    else: 
        continue
```

```python
game = []
for i in soup.select('.game_title > .game_link'):
    game.append(i.get_text())
```

```python
game
```

```python
#THIS ONE WORKS--MAIN DATA COLLECTION SCRIPT
gameid = []
games = []
descrips = []
authors = []
authorid = []
links = []
#for loop to get each game id from the attribute
for div in soup.select('div', class_='data-game_id'):
    if div.has_attr('data-game_id'):
        gameid.append(div['data-game_id'])
    else: 
        continue
#this one gets the title of the game
for i in soup.select('.game_title > a'):
    games.append(i.get_text())

#this one gets the description of the game
for i in soup.select('.game_text'):
    descrips.append(i.get_text())

#this one gets the author's name
for i in soup.select('.game_author'):
    authors.append(i.get_text())
    
#this one gets the author id
for div in soup.select('div > .game_author'):
    label = div.select("a")
    for userid in label:
        authorid.append(userid['data-label'])
        
#this one gets the link
for div in soup.select('div > .thumb_link'):
    if div.has_attr('href'):
        links.append(div['href'])
    else: 
        continue
```

```python
keys = ['gameid', 'title', 'description', 'author', 'authorid', 'price', 'saletag', 'link']
values = [gameid, game, descriptions, authors, authorid, prices, saletag, links]
```

```python
resdict = {keys[i]: values[i] for i in range(len(keys))}
```

```python
data = {'gameid': gameid, 'title': game, 'description': descriptions, 'author': authors, 'authorid': authorid, 'price': prices, 'saletag': saletag, 'link': links}
#reminder to check length of all arrays before saving to dictionary
```

```python
df = pd.DataFrame({ key:pd.Series(value) for key, value in resdict.items() })
```

```python
df
```

```python
df['authorid'] = df['authorid'].map(lambda x: x.lstrip('user:'))
```

```python
df.to_csv("itch.csv")
```
