```python
# Animal Crossing: New Horizons came out on 3/20/20, and basically everyone I know has been playing it.
# Since many places went into lockdown, a huge number of people have been having social gatherings in the game
# Some have even had graduations and birthday parties in the game to simulate physical gatherings.
# Because this game has become so important to people, I want to know what they're tweeting about.

htag = '%23AnimalCrossing'
n = 25000
acnh = twitter_api.search.tweets(q=htag, count=n)
acnh_statuses = acnh['statuses']
acnh_df = pd.read_json(json.dumps(acnh_statuses))
```
